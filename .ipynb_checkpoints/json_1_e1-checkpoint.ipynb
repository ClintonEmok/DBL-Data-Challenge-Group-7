```python
import numpy as np
import json as json
import pandas as pd
from pandas.io.json import json_normalize
```

# 1. Load Data

If you put the data set in the same folder as this notebook, you can use the following code to load the data set. 

```python
# Importing data from json file
df = pd.read_json('airlines-1.json', lines=True)
# Enabling option to show all columns
pd.set_option('display.max_columns', None)
# Filtering all tweet that are not labeled english
df_main = df.drop(df[df.lang != 'en'].index)
# Show the first  5 rows
df_main.head()
```

```python
# Created Flat dictionary of tweet data
df_main_flat= pd.json_normalize(json.loads(df_main.to_json(orient="records")))
df_main_flat
```

```python
# Calculate the amount of rows in dataframe before filtering on language
len(df.index)
# Calculate the amount of rows in dataframe after filtering on language
len(df_main.index)

# Calculate the amount of columns in dataframe 
len(df_main.columns)

# Calculate the amount of columns in dataframe when flattened
len(df_main_flat.columns)
```

The json file contains several nested objects of which the most interesting are:
* *language (lang)*: indicates the language of the tweet
* *user*: contains information regarding a user such as id, location and verification status
* *entities*: contains information regarding a tweet such as hashtags and user mentions
