```python
text_list = df_main['text'].tolist()
text_list
Bal_list = [t for t in text_list if 'British Airways' in t] + [y for y in text_list if 'British_Airways' in y]
nRT_Bal_list = [u for u in Bal_list if 'RT' not in u]
nRT_Bal_list
KLM_list = [i for i in text_list if 'KLM' in i]
nRT_KLM_list = [o for o in KLM_list if 'RT' not in o]
nRT_KLM_list
AAl_list = [t for t in text_list if 'American Airlines' in t] + [y for y in text_list if 'AmericanAir' in y]
nRT_AAl_list = [u for u in AAl_list if 'RT' not in u]
nRT_AAl_list
print('KLM', len(nRT_KLM_list), 'B-airlines', len(nRT_Bal_list), 'A-airlines', len(nRT_AAl_list))
```
